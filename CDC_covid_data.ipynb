## United States COVID-19 Cases and Deaths by State over Time
Source: https://healthdata.gov/dataset/United-States-COVID-19-Cases-and-Deaths-by-State-o/hiyb-zgc2

```python
df_case = pd.read_csv('/content/drive/MyDrive/group_9/Datasets/COVID_related/United_States_COVID-19_Cases_and_Deaths_by_State_over_Time.csv', parse_dates = ['submission_date'])
df_case.set_index('submission_date', inplace=True, drop=False)
```

```python
df_case.head(2)
```

```python
#All states in 1 chart
for state in df_case.state.unique():
  state_df = df_case.loc[df_case['state']==state, :]
  state_df.groupby(pd.Grouper(freq='M'))["new_case"].mean().plot()
```

```python
#All states in 1 chart
for state in df_case.state.unique():
  state_df = df_case.loc[df_case['state']==state, :]
  state_df.groupby(pd.Grouper(freq='M'))["new_death"].mean().plot()
  # plt.show()
```

```python
for state in df_case.state.unique():
  state_df = df_case.loc[df_case['state']==state, :]
  state_df.groupby(pd.Grouper(freq='M'))["tot_cases"].mean().plot()
  # plt.show()
```
