```python
df_vac = pd.read_csv('/content/drive/MyDrive/DS4A/Datasets /COVID_related/COVID-19_Vaccinations_in_the_United_States_Jurisdiction.csv')
```

```python
df_vac.reset_index(drop=True, inplace=True)
df_vac.head(2)
```

```python
#drop week column
df_agg = df_vac.drop(columns='MMWR_week')
```

```python
df_agg.head(2)
```

```python
df_agg.Date = pd.to_datetime(df_agg.Date) 
df_agg.head(2)
```

```python
# group by state then get the monthly mean with resample
# since this is grouping by the mean and not the count, we can only use the Pct (%) features! 
df = df_agg.groupby('Location').resample('M', on = "Date").mean()
```
